```python
import pandas as pd
import numpy as np
import random
import os
import time
import json
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from accelerate import Accelerator
# Evaluation metrics
from sacrebleu import corpus_bleu
from sacrebleu.metrics import CHRF
from bleurt import score as bleurt_score

# For API access
import openai
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# Google PaLM/Gemini API
import google.generativeai as genai


class TranslationEvaluator:
    def __init__(self, train_df, test_df, source_lang='ba', target_lang='en',
                 few_shot_count=10, seed=42, cache_dir='./translation_cache'):
        """
        Initialize the translation evaluator.

        Args:
            train_df: DataFrame with source_lang and target_lang columns
            test_df: DataFrame with source_lang and target_lang columns
            source_lang: Source language column name
            target_lang: Target language column name
            few_shot_count: Number of examples to use in few-shot prompting
            seed: Random seed for reproducibility
            cache_dir: Directory to cache translation results
        """
        self.train_df = train_df
        self.test_df = test_df
        self.source_lang = source_lang
        self.target_lang = target_lang
        self.few_shot_count = few_shot_count
        self.seed = seed
        self.cache_dir = cache_dir

        # Create cache directory if it doesn't exist
        if not os.path.exists(cache_dir):
            os.makedirs(cache_dir)

        # Set random seed for reproducibility
        random.seed(self.seed)
        np.random.seed(self.seed)

        # Sample few-shot examples
        self.few_shot_examples = self._sample_few_shot_examples()

        # Dictionary to store translation results
        self.results = {}

    def _sample_few_shot_examples(self):
        """Sample few-shot examples from the training data."""
        return self.train_df.sample(self.few_shot_count)

    def construct_prompt(self, input_text, model_name="gpt"):
        """
        Construct a prompt for translation with few-shot examples.

        Args:
            input_text: Text to translate
            model_name: Type of model to format the prompt for

        Returns:
            Formatted prompt string
        """
        source_lang_name = 'Bahnar' if self.source_lang == 'ba' else 'English'
        target_lang_name = 'English' if self.target_lang == 'en' else 'Bahnar'

        if model_name in ["gpt-3.5", "gpt-4"]:
            # Format for OpenAI models
            prompt = f"Translate the following text from {source_lang_name} to {target_lang_name}.\n\n"
            prompt += "Here are some examples:\n\n"

            for _, row in self.few_shot_examples.iterrows():
                prompt += f"{source_lang_name}: {row[self.source_lang]}\n"
                prompt += f"{target_lang_name}: {row[self.target_lang]}\n\n"

            prompt += f"{source_lang_name}: {input_text}\n"
            prompt += f"{target_lang_name}: "

        elif model_name in ["gemini", "palm"]:
            # Format for Google PaLM/Gemini models
            prompt = f"Translate the following text from {source_lang_name} to {target_lang_name}.\n\n"
            prompt += "Here are some examples:\n\n"

            for _, row in self.few_shot_examples.iterrows():
                prompt += f"{source_lang_name}: {row[self.source_lang]}\n"
                prompt += f"{target_lang_name}: {row[self.target_lang]}\n\n"

            prompt += f"{source_lang_name}: {input_text}\n"
            prompt += f"{target_lang_name}: "

        else:
            # Format for open-source models
            prompt = f"Translate the following text from {source_lang_name} to {target_lang_name}.\n\n"
            prompt += "Here are some examples:\n\n"

            for _, row in self.few_shot_examples.iterrows():
                prompt += f"{source_lang_name}: {row[self.source_lang]}\n"
                prompt += f"{target_lang_name}: {row[self.target_lang]}\n\n"

            prompt += f"{source_lang_name}: {input_text}\n"
            prompt += f"{target_lang_name}: "

        return prompt

    def translate_with_openai(self, model_name, api_key, batch_size=10, max_retries=3):
            """
            Translate test data using OpenAI API, with version-agnostic implementation.

            Args:
                model_name: Name of the OpenAI model to use (e.g., "gpt-3.5-turbo", "gpt-4")
                api_key: OpenAI API key
                batch_size: Number of examples to process at once
                max_retries: Maximum number of retry attempts on API failures

            Returns:
                List of dictionaries with original text, reference translation, and model translation
            """
            import os # Import os inside the method
            import json # Import json inside the method
            import importlib.metadata # Import importlib.metadata inside the method
            import requests # Import requests inside the method
            import time # Import time inside the method

            # Check if we already have cached results
            cache_file = os.path.join(self.cache_dir, f"{model_name}_translations.json")
            if os.path.exists(cache_file):
                print(f"Loading cached translations for {model_name}...")
                with open(cache_file, 'r', encoding='utf-8') as f:
                    translations = json.load(f)
                    # Store in results dictionary
                    self.results[model_name] = translations
                    return translations

            # Determine OpenAI library version and setup accordingly
            import openai

            try:
                openai_version = importlib.metadata.version('openai')
                is_old_version = int(openai_version.split('.')[0]) < 1
            except:
                # If we can't determine the version, try to detect based on API structure
                is_old_version = hasattr(openai, 'ChatCompletion')

            print(f"Using OpenAI library version: {'0.x' if is_old_version else '1.x+'}")

            if is_old_version:
                # Old version (pre-1.0.0)
                openai.api_key = api_key
            else:
                # New version (1.0.0+)
                try:
                    # Try importing the required components
                    # import os # os is already imported at the top of the method
                    os.environ["OPENAI_API_KEY"] = api_key
                except Exception as e:
                    print(f"Error setting up OpenAI client: {e}")

            print(f"Translating {len(self.test_df)} examples with {model_name}...")
            translations = []

            for i in tqdm(range(0, len(self.test_df), batch_size)):
                batch = self.test_df.iloc[i:i+batch_size]

                for _, row in batch.iterrows():
                    source_text = row[self.source_lang]

                    # Construct the prompt
                    prompt = self.construct_prompt(source_text, model_name)

                    # Try to get translation with retries
                    for attempt in range(max_retries):
                        try:
                            if is_old_version:
                                # Old version API call
                                response = openai.ChatCompletion.create(
                                    model=model_name,
                                    messages=[
                                        {"role": "user", "content": prompt}
                                    ],
                                    temperature=0.3,
                                    max_tokens=150
                                )
                                translation = response['choices'][0]['message']['content'].strip()
                            else:
                                # New version API call using lower-level requests
                                # import requests # requests is already imported at the top of the method
                                # import json # json is already imported at the top of the method

                                headers = {
                                    "Content-Type": "application/json",
                                    "Authorization": f"Bearer {api_key}"
                                }

                                payload = {
                                    "model": model_name,
                                    "messages": [{"role": "user", "content": prompt}],
                                    "temperature": 0.3,
                                    "max_tokens": 150
                                }

                                response = requests.post(
                                    "https://api.openai.com/v1/chat/completions",
                                    headers=headers,
                                    json=payload
                                )

                                if response.status_code == 200:
                                    data = response.json()
                                    translation = data['choices'][0]['message']['content'].strip()
                                else:
                                    raise Exception(f"API request failed with status {response.status_code}: {response.text}")

                            translations.append({
                                'source': source_text,
                                'reference': row[self.target_lang],
                                'translation': translation
                            })

                            # Small delay to avoid rate limiting
                            time.sleep(0.5)
                            break
                        except Exception as e:
                            print(f"Error with {model_name} (attempt {attempt+1}/{max_retries}): {e}")
                            if attempt == max_retries - 1:
                                # If it's the last attempt, add a placeholder
                                translations.append({
                                    'source': source_text,
                                    'reference': row[self.target_lang],
                                    'translation': "[ERROR: Translation failed]"
                                })
                            time.sleep(2 ** attempt)  # Exponential backoff

            # Cache the results
            with open(cache_file, 'w', encoding='utf-8') as f:
                json.dump(translations, f, ensure_ascii=False, indent=2)

            # Store in results dictionary
            self.results[model_name] = translations

            return translations

    def translate_with_google_model(self, model_name, api_key, model_id="gemini-1.5-pro-latest", batch_size=10, max_retries=3):
        """
        Translate test data using Google's PaLM/Gemini API.

        Args:
            model_name: Name to use for this model in result tracking
            api_key: Google API key
            model_id: Model ID to use (e.g., "gemini-pro")
            batch_size: Number of examples to process at once
            max_retries: Maximum number of retry attempts on API failures

        Returns:
            DataFrame with original text and translations
        """
        # Check if we already have cached results
        cache_file = os.path.join(self.cache_dir, f"{model_name}_translations.json")
        if os.path.exists(cache_file):
            print(f"Loading cached translations for {model_name}...")
            with open(cache_file, 'r', encoding='utf-8') as f:
                translations = json.load(f)
            return translations

        # Configure the API
        genai.configure(api_key=api_key)

        # Get the model
        model = genai.GenerativeModel(model_id)

        print(f"Translating {len(self.test_df)} examples with {model_name}...")
        translations = []

        for i in tqdm(range(0, len(self.test_df), batch_size)):
            batch = self.test_df.iloc[i:i+batch_size]

            for _, row in batch.iterrows():
                source_text = row[self.source_lang]

                # Construct the prompt
                prompt = self.construct_prompt(source_text, model_name)

                # Try to get translation with retries
                for attempt in range(max_retries):
                    try:
                        # Generate content
                        response = model.generate_content(prompt)

                        # Extract translation
                        if hasattr(response, 'text'):
                            translation = response.text.strip()
                        else:
                            # For older API versions that return different response structure
                            translation = response.candidates[0].content.parts[0].text.strip()

                        translations.append({
                            'source': source_text,
                            'reference': row[self.target_lang],
                            'translation': translation
                        })

                        # Small delay to avoid rate limiting
                        time.sleep(0.5)
                        break
                    except Exception as e:
                        print(f"Error with {model_name} (attempt {attempt+1}/{max_retries}): {e}")
                        if attempt == max_retries - 1:
                            # If it's the last attempt, add a placeholder
                            translations.append({
                                'source': source_text,
                                'reference': row[self.target_lang],
                                'translation': "[ERROR: Translation failed]"
                            })
                        time.sleep(2 ** attempt)  # Exponential backoff

        # Cache the results
        with open(cache_file, 'w', encoding='utf-8') as f:
            json.dump(translations, f, ensure_ascii=False, indent=2)

        # Store in results dictionary
        self.results[model_name] = translations

        return translations

    def translate_with_hf_model(self, model_name, model_path_or_name, batch_size=4, max_length=150):
        """
        Translate test data using Hugging Face models (like LLaMA, Mistral, Mixtral).

        Args:
            model_name: Name to use for this model in result tracking
            model_path_or_name: Hugging Face model path or name
            batch_size: Number of examples to process at once
            max_length: Maximum output sequence length

        Returns:
            DataFrame with original text and translations
        """
        # Check if we already have cached results
        accelerator = Accelerator()
        cache_file = os.path.join(self.cache_dir, f"{model_name}_translations.json")
        if os.path.exists(cache_file):
            print(f"Loading cached translations for {model_name}...")
            with open(cache_file, 'r', encoding='utf-8') as f:
                translations = json.load(f)
            return translations

        print(f"Loading {model_name} from {model_path_or_name}...")

        # Load model and tokenizer
        tokenizer = AutoTokenizer.from_pretrained(model_path_or_name)
        tokenizer.pad_token = tokenizer.eos_token
        try:
            # For smaller models (7B and below)
            model = AutoModelForCausalLM.from_pretrained(
                model_path_or_name,
                # torch_dtype=torch.float16 if device == "cuda" else torch.float32,
                device_map="auto"
            )

            # Use pipeline for easier inference
            generator = pipeline(
                "text-generation",
                model=model,
                tokenizer=tokenizer,
                # device=device
            )
        except Exception as e:
            print(f"Failed to load model fully into memory: {e}")
            print("Using 8-bit quantization...")
            # For larger models (using 8-bit quantization)
            model = AutoModelForCausalLM.from_pretrained(
                model_path_or_name,
                # torch_dtype=torch.float16 if device == "cuda" else torch.float32,
                device_map="auto",
                load_in_8bit=True
            )

            # Use pipeline for easier inference
            generator = pipeline(
                "text-generation",
                model=model,
                tokenizer=tokenizer,
                # device=device
            )

        print(f"Translating {len(self.test_df)} examples with {model_name}...")
        translations = []

        for i in tqdm(range(0, len(self.test_df), batch_size)):
            batch = self.test_df.iloc[i:i+batch_size]

            for _, row in batch.iterrows():
                source_text = row[self.source_lang]

                # Construct the prompt
                prompt = self.construct_prompt(source_text, model_name)

                try:
                    # Generate translation
                    output = generator(
                        prompt,
                        max_new_tokens=max_length,
                        do_sample=False,
                        temperature=0.3,
                        num_return_sequences=1
                    )

                    generated_text = output[0]['generated_text']

                    # Extract the translation from the generated text
                    translation = generated_text[len(prompt):]

                    # Clean up any trailing text or format issues
                    translation = translation.strip().split('\n')[0].strip()

                    translations.append({
                        'source': source_text,
                        'reference': row[self.target_lang],
                        'translation': translation
                    })

                except Exception as e:
                    print(f"Error with {model_name}: {e}")
                    translations.append({
                        'source': source_text,
                        'reference': row[self.target_lang],
                        'translation': "[ERROR: Translation failed]"
                    })

        # Cache the results
        with open(cache_file, 'w', encoding='utf-8') as f:
            json.dump(translations, f, ensure_ascii=False, indent=2)

        # Store in results dictionary
        self.results[model_name] = translations

        return translations

    def calculate_metrics(self, translations):
        references = [[t['reference']] for t in translations]
        hypothesis = [t['translation'] for t in translations]

        # BLEU & chrF
        bleu = corpus_bleu(hypothesis, references)
        chrf = CHRF()
        chrf_score = chrf.corpus_score(hypothesis, references)

        # BLEURT (optional, requires BLEURT setup)
        bleurt_scorer = bleurt_score.BleurtScorer("BLEURT-20")
        bleurt_scores = bleurt_scorer.score(references=[r[0] for r in references], candidates=hypothesis)
        avg_bleurt = sum(bleurt_scores) / len(bleurt_scores)

        return {
            'bleu': bleu.score,
            'chrf': chrf_score.score,
            'bleurt': avg_bleurt
        }

    def evaluate_all(self):
        """
        Evaluate all translation results and summarize metrics.

        Returns:
            DataFrame with model names and their respective metrics
        """
        evaluation_results = []

        for model_name, translations in self.results.items():
            metrics = self.calculate_metrics(translations)

            evaluation_results.append({
                'model': model_name,
                'bleu': metrics['bleu'],
                'chrf': metrics['chrf'],
                'bleurt': metrics['bleurt'],
                'sample_count': len(translations)
            })

        # Convert to DataFrame for easier visualization
        return pd.DataFrame(evaluation_results)

    def visualize_results(self, results_df):
        """
        Visualize the evaluation results.

        Args:
            results_df: DataFrame with model names and metrics
        """
        # Set the style
        plt.style.use('ggplot')

        # Set up the figure
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

        # Plot BLEU scores
        sns.barplot(x='model', y='bleu', data=results_df, ax=ax1)
        ax1.set_title('BLEU Scores by Model')
        ax1.set_ylabel('BLEU Score')
        ax1.set_xlabel('Model')

        # Plot chrF scores
        sns.barplot(x='model', y='chrf', data=results_df, ax=ax2)
        ax2.set_title('chrF Scores by Model')
        ax2.set_ylabel('chrF Score')
        ax2.set_xlabel('Model')

        # Adjust layout
        plt.tight_layout()
        plt.savefig('translation_metrics.png', dpi=300)
        plt.show()

    def save_example_translations(self, num_examples=5):
        """
        Save a few example translations from each model for qualitative analysis.

        Args:
            num_examples: Number of examples to save

        Returns:
            DataFrame with example translations
        """
        examples_df = pd.DataFrame()

        # Select a few random examples
        indices = random.sample(range(len(self.test_df)), num_examples)

        for i, idx in enumerate(indices):
            example = {
                'Example': i+1,
                'Source': self.test_df.iloc[idx][self.source_lang],
                'Reference': self.test_df.iloc[idx][self.target_lang]
            }

            # Add translations from each model
            for model_name, translations in self.results.items():
                if idx < len(translations):
                    example[f'{model_name}'] = translations[idx]['translation']
                else:
                    example[f'{model_name}'] = 'N/A'

            examples_df = pd.concat([examples_df, pd.DataFrame([example])], ignore_index=True)

        # Save to CSV
        model_names = '_'.join(self.results.keys())
        examples_df.to_csv(f'/content/drive/MyDrive/model_result/translations_{model_names}.csv', index=False)

        return examples_df
```

```python
class LLama2TranslationEvaluator:
    def __init__(self, train_df, test_df, source_lang='ba', target_lang='en',
                 few_shot_count=10, seed=42, cache_dir='./translation_cache'):
        self.train_df = train_df
        self.test_df = test_df
        self.source_lang = source_lang
        self.target_lang = target_lang
        self.few_shot_count = few_shot_count
        self.seed = seed
        self.cache_dir = cache_dir

        os.makedirs(cache_dir, exist_ok=True)
        random.seed(self.seed)
        np.random.seed(self.seed)
        self.few_shot_examples = self.train_df.sample(self.few_shot_count)
        self.results = {}

    def construct_prompt(self, input_text, model_name="gpt"):
        lang_map = {'ba': 'Bahnar', 'en': 'English'}
        source_lang_name = lang_map.get(self.source_lang, self.source_lang)
        target_lang_name = lang_map.get(self.target_lang, self.target_lang)

        prompt_parts = [f"Translate the following text from {source_lang_name} to {target_lang_name}.\n"]
        for _, row in self.few_shot_examples.iterrows():
            prompt_parts.append(f"{source_lang_name}: {row[self.source_lang]}\n{target_lang_name}: {row[self.target_lang]}\n")
        prompt_parts.append(f"{source_lang_name}: {input_text}\n{target_lang_name}: ")

        return "\n".join(prompt_parts)

    def translate_with_google_model(self, model_name, api_key, model_id="gemini-1.5-pro-latest", batch_size=10, max_retries=3):
        cache_file = os.path.join(self.cache_dir, f"{model_name}_translations.json")
        if os.path.exists(cache_file):
            with open(cache_file, 'r', encoding='utf-8') as f:
                return json.load(f)

        genai.configure(api_key=api_key)
        model = genai.GenerativeModel(model_id)
        translations = []

        for _, row in tqdm(self.test_df.iterrows(), total=len(self.test_df)):
            prompt = self.construct_prompt(row[self.source_lang], model_name)
            for attempt in range(max_retries):
                try:
                    response = model.generate_content(prompt)
                    translation = response.text.strip() if hasattr(response, 'text') else response.candidates[0].content.parts[0].text.strip()
                    translations.append({
                        'source': row[self.source_lang],
                        'reference': row[self.target_lang],
                        'translation': translation
                    })
                    break
                except Exception as e:
                    print(f"[{model_name}] Attempt {attempt+1} failed: {e}")
                    if attempt == max_retries - 1:
                        translations.append({
                            'source': row[self.source_lang],
                            'reference': row[self.target_lang],
                            'translation': "[ERROR: Translation failed]"
                        })
                    time.sleep(2 ** attempt)

        with open(cache_file, 'w', encoding='utf-8') as f:
            json.dump(translations, f, ensure_ascii=False, indent=2)

        self.results[model_name] = translations
        return translations

    def translate_with_hf_model(self, model_name, model_path_or_name, batch_size=4, max_length=150):
        cache_file = os.path.join(self.cache_dir, f"{model_name}_translations.json")
        if os.path.exists(cache_file):
            with open(cache_file, 'r', encoding='utf-8') as f:
                return json.load(f)

        tokenizer = AutoTokenizer.from_pretrained(model_path_or_name)
        tokenizer.pad_token = tokenizer.eos_token

        try:
            model = AutoModelForCausalLM.from_pretrained(model_path_or_name, device_map="auto")
        except Exception:
            model = AutoModelForCausalLM.from_pretrained(model_path_or_name, device_map="auto", load_in_8bit=True)

        generator = pipeline("text-generation", model=model, tokenizer=tokenizer)
        translations = []

        for _, row in tqdm(self.test_df.iterrows(), total=len(self.test_df)):
            prompt = self.construct_prompt(row[self.source_lang], model_name)
            try:
                output = generator(prompt, max_new_tokens=max_length, do_sample=False, temperature=0.3)
                translation = output[0]['generated_text'][len(prompt):].strip().split('\n')[0]
                translations.append({
                    'source': row[self.source_lang],
                    'reference': row[self.target_lang],
                    'translation': translation
                })
            except Exception as e:
                print(f"[{model_name}] Error: {e}")
                translations.append({
                    'source': row[self.source_lang],
                    'reference': row[self.target_lang],
                    'translation': "[ERROR: Translation failed]"
                })

        with open(cache_file, 'w', encoding='utf-8') as f:
            json.dump(translations, f, ensure_ascii=False, indent=2)

        self.results[model_name] = translations
        return translations

    def calculate_metrics(self, translations):
        references = [[t['reference']] for t in translations]
        hypothesis = [t['translation'] for t in translations]

        bleu = corpus_bleu(hypothesis, references).score
        chrf = CHRF().corpus_score(hypothesis, references).score
        bleurt_scorer = bleurt_score.BleurtScorer("BLEURT-20")
        bleurt_scores = bleurt_scorer.score(references=[r[0] for r in references], candidates=hypothesis)
        avg_bleurt = sum(bleurt_scores) / len(bleurt_scores)

        return {'bleu': bleu, 'chrf': chrf, 'bleurt': avg_bleurt}

    def evaluate_all(self):
        evaluation_results = [
            {'model': model_name, **self.calculate_metrics(trans)}
            for model_name, trans in self.results.items()
        ]
        return pd.DataFrame(evaluation_results)

    def visualize_results(self, results_df):
        plt.style.use('ggplot')
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
        sns.barplot(x='model', y='bleu', data=results_df, ax=ax1)
        ax1.set_title('BLEU Scores by Model')
        sns.barplot(x='model', y='chrf', data=results_df, ax=ax2)
        ax2.set_title('chrF Scores by Model')
        plt.tight_layout()
        plt.savefig('translation_metrics.png', dpi=300)
        plt.show()

    def save_example_translations(self, num_examples=5):
        indices = random.sample(range(len(self.test_df)), num_examples)
        examples = []
        for i, idx in enumerate(indices):
            row = {
                'Example': i+1,
                'Source': self.test_df.iloc[idx][self.source_lang],
                'Reference': self.test_df.iloc[idx][self.target_lang]
            }
            for model_name, trans in self.results.items():
                row[model_name] = trans[idx]['translation'] if idx < len(trans) else 'N/A'
            examples.append(row)

        df = pd.DataFrame(examples)
        out_path = os.path.join(self.cache_dir, f'translations_{"_".join(self.results.keys())}.csv')
        df.to_csv(out_path, index=False)
        return df
```
